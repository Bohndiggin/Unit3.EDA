- mean
> mean is the average of all the numbers
- median
> median is the middle of all the numbers.
- mode
> mode is the most common value in the data.
- central tendency
> the central or typical value it can be mean median or mode.
- distribution
> Distribution is a mathematical function that gives probabilities of outcomes. AKA how likely is any number to be in this segment of the data.
- variance
> how spread out is the data from the mean. how 'wide' is the bell curve/shape.
- standard deviation
> like variance, but mathmatically different. square root not square.

### Step 3 - Use Pearsonr & Spearmanr To Answer Questions About Correlation and P-Values

<h3 style='color:red;'> First, load 'salary_numeric.csv' - where education has been re-expressed as numeric variables.</h3>

```python
# load salary_numeric.csv into dataframe where education levels have been re-expressed as numeric

df1 = pd.read_csv('salary_numeric.csv')
```

```python
################################### TO DO #########################################
####### FOLLOW PROJECT INSTRUCTIONS TO COMPLETE THIS STEP IN THE CELLS BELOW ######
```

```python
# import pearson and spearmanr methods from scipy.stats library

from scipy.stats import pearsonr, spearmanr
```

### What is the correlation between salary and education?

#### Your Hypothesis = Higher Education leads to higher salary with diminishing returns.

```python
# calculate correlation and pvalue of salary and education with spearmanr method

spearmanr(df1.salary, df1.education)
```

Medium amount of correlation, tiny p-value

#### Are these results statistically significant? How do you know?

The tiny P-value indicates that they are related, but the slight negative correlation shows that it's not too large of an impact.

### What is the correlation between salary and age?

#### Your Hypothesis Older=Higher pay.

```python
# calculate correlation and pvalue of salary and age with pearsonr method

pearsonr(df1.salary, df1.age)
```

higher correlation.

#### Are these results statistically significant? How do you know? 

Absolutly significant related and in a bigish way

### What is the correlation between salary and industry experience? 

#### Your Hypothesis High experience means higher pay

```python
# calculate correlation and pvalue of salary and industry experience with pearsonr 

pearsonr(df1.salary, df1.industry_experience)
```

High! But not as much as age

#### Are these results statistically significant? How do you know?

Yes. Low P, highish effect

#### Crosstab of average salary grouped by education and industry

```python
# use pd.crosstab() to calculate mean salary in each industry grouped by education level

pd.crosstab(index=df1.industry, columns=df1.education, values=df1.salary, aggfunc=np.mean).round(0)
```

#### Heatmap of crosstab of average salary grouped by education and industry

```python
# save above result to new variable

tab = pd.crosstab(index=df1.industry, columns=df1.education, values=df1.salary, aggfunc=np.mean).round(0)
```

```python
# create heatmap with annotated mean salary in each industry and education level

sns.heatmap(tab, annot=True)
```

Education impacts finance and tech more than business and healthcare
